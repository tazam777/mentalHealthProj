```python
# Function to clean gender values more inclusively
def clean_gender(gender):
    gender = gender.strip().lower()  # Normalize string by stripping spaces and lowering case
    if gender in ['male', 'm', 'man', 'cis male', 'male (cis)', 'male ']:
        return 'Male'
    elif gender in ['female', 'f', 'woman', 'cis female', 'female ', 'femake', 'femail']:
        return 'Female'
    elif gender in ['trans male', 'trans man', 'ftm', 'trans-male']:
        return 'Trans Male'
    elif gender in ['trans female', 'trans woman', 'mtf', 'trans-female']:
        return 'Trans Female'
    elif gender in ['non-binary', 'genderqueer', 'androgyne', 'agender', 'enby', 'fluid', 'neuter', 
                    'queer', 'queer/she/they', 'male leaning androgynous', 'something kinda male?', 'genderqueer']:
        return 'Non-binary/Other'
    else:
        return 'Non-binary/Other'  # Default category for unclear/ambiguous gender identities

# Apply the function to the Gender column
df['Gender'] = df['Gender'].apply(clean_gender)

# Check the unique values after cleaning
print("Unique values in Gender column after cleaning:")
print(df['Gender'].unique())
```

```python
# Import pycountry for standardizing country names
import pycountry

# Function to standardize country names using pycountry
def standardize_country(country_name):
    """
    Standardize country names using the pycountry library.
    
    Args:
        country_name (str): Raw country name.
        
    Returns:
        str: Standardized country name or original if not found.
    """
    # Normalize case
    country_name = country_name.strip().title()  # Ensures 'brazil' -> 'Brazil'

    # Try to look up the country using pycountry
    try:
        return pycountry.countries.lookup(country_name).name
    except LookupError:
        # Return original name if not found in pycountry
        return country_name


# Apply the function to the 'Country' column
df['Country'] = df['Country'].apply(standardize_country)

# Checking unique values after standardization
print("Unique values in Country column after standardization:")
print(df['Country'].unique())
```

```python
# Import necessary libraries
import os
import joblib  # For saving the model and other objects
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score, classification_report
from sklearn.model_selection import train_test_split, StratifiedKFold, cross_val_score
from sklearn.preprocessing import LabelEncoder, StandardScaler
import pandas as pd
import numpy as np
import random
from collections import Counter

# Step 0: Reset everything (ensure no previous state is lingering)
import gc
gc.collect()  # Collect garbage from previous runs (helps clear environment)
SEED = 42
np.random.seed(SEED)
random.seed(SEED)

# Step 1: Create a backup of the data before any target or transformations
df_backup = df.copy()  # Backup original data

# Step 2: Create the target column for the employee model
df['mental_health_friendly_score'] = df.apply(
    lambda row: 1 if (row['benefits'] == 'Yes' or row['care_options'] == 'Yes' or row['anonymity'] == 'Yes') else 0, axis=1
)

# Step 3: Define features for the employee model
employee_features = ['self_employed', 'work_interfere', 'remote_work', 
                     'family_history', 'benefits', 'leave', 'care_options', 'treatment']

# Apply Label Encoding to convert categorical features into numeric values
label_encoders = {}
for col in employee_features:
    le = LabelEncoder()
    df[col] = le.fit_transform(df[col].astype(str))
    label_encoders[col] = le

# Feature Engineering
df['policy_support_score'] = df[['benefits', 'care_options', 'leave']].sum(axis=1)
df['self_employed_remote_work'] = df['self_employed'] * df['remote_work']
df['mental_health_history_score'] = df[['family_history']].sum(axis=1)
df['remote_work_leave_ratio'] = df['remote_work'] / (df['leave'] + 1)
df['benefits_leave_interaction'] = df['benefits'] * df['leave']
df['care_options_work_interaction'] = df['care_options'] * df['work_interfere']

employee_features_new = employee_features + ['policy_support_score', 'self_employed_remote_work', 
                                             'mental_health_history_score', 'remote_work_leave_ratio', 
                                             'benefits_leave_interaction', 'care_options_work_interaction']

# Split the data into training and testing sets
X_train_emp, X_test_emp, y_train_emp, y_test_emp = train_test_split(
    df[employee_features_new], df['mental_health_friendly_score'], test_size=0.2, random_state=SEED, stratify=df['mental_health_friendly_score']
)

# Scale the data
scaler = StandardScaler()
X_train_emp_scaled = scaler.fit_transform(X_train_emp)
X_test_emp_scaled = scaler.transform(X_test_emp)

# Train a Random Forest Classifier
rf_model = RandomForestClassifier(random_state=SEED)
rf_model.fit(X_train_emp_scaled, y_train_emp)

# Evaluate the model
y_pred_rf = rf_model.predict(X_test_emp_scaled)
print(f"Employee Model Accuracy (Random Forest): {accuracy_score(y_test_emp, y_pred_rf):.2f}")
print("Classification Report (Random Forest):")
print(classification_report(y_test_emp, y_pred_rf))

# Cross-validation
kf = StratifiedKFold(n_splits=5, shuffle=True, random_state=SEED)
cv_scores = cross_val_score(rf_model, X_train_emp_scaled, y_train_emp, cv=kf, scoring='accuracy')
print(f"Cross-validation accuracy scores (Random Forest): {cv_scores}")
print(f"Mean cross-validation accuracy (Random Forest): {np.mean(cv_scores):.2f}")

# Save the model and preprocessing tools
os.makedirs("models", exist_ok=True)  # Create 'models' directory if it doesn't exist

joblib.dump(rf_model, "models/employee_model.pkl")  # Save the trained model
joblib.dump(scaler, "models/employee_scaler.pkl")  # Save the scaler
joblib.dump(label_encoders, "models/employee_label_encoders.pkl")  # Save the label encoders

print("Model and preprocessing tools have been saved in the 'models' folder.")
```

```python
# Import necessary libraries
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score, classification_report
from sklearn.model_selection import train_test_split, StratifiedKFold, cross_val_score
from sklearn.preprocessing import LabelEncoder, StandardScaler
import pandas as pd
import numpy as np
import random
from collections import Counter
import joblib
import os

# Step 1: Set random seed for reproducibility
SEED = 42
np.random.seed(SEED)
random.seed(SEED)

# Step 2: Define company-specific features from the available columns
company_features = ['benefits', 'wellness_program', 'anonymity', 'leave', 
                    'seek_help', 'remote_work', 'mental_health_consequence', 'supervisor']

# Step 3: Create the target column for the company model
df['mental_health_friendly_score'] = df.apply(
    lambda row: 1 if (
        row['benefits'] == 'Yes' or 
        row['wellness_program'] == 'Yes' or 
        row['anonymity'] == 'Yes' or
        row['leave'] in ['Very easy', 'Somewhat easy'] or
        row['seek_help'] == 'Yes') else 0, axis=1
)

# Debug: Check the target variable distribution
print("Target variable distribution (mental_health_friendly_score):")
print(df['mental_health_friendly_score'].value_counts())

# Step 4: Apply Label Encoding to convert categorical features into numeric values
label_encoders = {}
for col in company_features:
    le = LabelEncoder()
    df[col] = le.fit_transform(df[col].astype(str))  # Convert all to string before encoding
    label_encoders[col] = le  # Store the label encoder if needed later

# Debug: Check if the data was encoded properly
print("Data after label encoding:")
print(df[company_features].head())

# Step 5: Split the data into training and testing sets (with stratification on the target variable)
X_train_comp, X_test_comp, y_train_comp, y_test_comp = train_test_split(
    df[company_features], df['mental_health_friendly_score'], test_size=0.2, random_state=SEED, stratify=df['mental_health_friendly_score']
)

# Debugging: Check class distribution in train and test
print(f"Train/Test split - y_train class distribution: {Counter(y_train_comp)}")
print(f"Train/Test split - y_test class distribution: {Counter(y_test_comp)}")
print(f"Train shape: {X_train_comp.shape}, Test shape: {X_test_comp.shape}")

# Step 6: Scale the training and testing sets separately
scaler = StandardScaler()
X_train_comp_scaled = scaler.fit_transform(X_train_comp)  # Scale training data
X_test_comp_scaled = scaler.transform(X_test_comp)  # Scale test data

# Step 7: Train a Random Forest Classifier with constraints to reduce overfitting
rf_model = RandomForestClassifier(random_state=SEED, n_estimators=100, max_depth=5)
rf_model.fit(X_train_comp_scaled, y_train_comp)

# Test the model and evaluate performance
y_pred_rf = rf_model.predict(X_test_comp_scaled)
print(f"Company Model Accuracy (Random Forest): {accuracy_score(y_test_comp, y_pred_rf):.2f}")
print("Classification Report (Random Forest):")
print(classification_report(y_test_comp, y_pred_rf))

# Step 8: Cross-validation using Random Forest (StratifiedKFold)
kf = StratifiedKFold(n_splits=5, shuffle=True, random_state=SEED)
cv_scores = cross_val_score(rf_model, X_train_comp_scaled, y_train_comp, cv=kf, scoring='accuracy')
print(f"Cross-validation accuracy scores (Random Forest): {cv_scores}")
print(f"Mean cross-validation accuracy (Random Forest): {np.mean(cv_scores):.2f}")

# Step 9: Save the model, scaler, and label encoders

# Create a directory for saving models if it doesn't exist
os.makedirs('models', exist_ok=True)

# Save the trained Random Forest model
joblib.dump(rf_model, 'models/company_model.pkl')
print("Company Model saved as 'models/company_model.pkl'.")

# Save the scaler used for data preprocessing
joblib.dump(scaler, 'models/company_scaler.pkl')
print("Scaler saved as 'models/company_scaler.pkl'.")

# Save label encoders for each categorical feature
for col, le in label_encoders.items():
    joblib.dump(le, f'models/{col}_label_encoder.pkl')
    print(f"Label Encoder for '{col}' saved as 'models/{col}_label_encoder.pkl'.")
```
